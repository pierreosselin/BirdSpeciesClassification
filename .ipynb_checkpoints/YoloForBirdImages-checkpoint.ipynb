### Cell cropping and saving images

```python
def crop_images(path, model):
    image = Image.open(path)
    out_boxes, out_scores, out_classes = model.detect_image(image)
    flag = False
    for i in range(len(out_boxes)):
        if out_classes[i] == 14:
            if out_scores[i] <= 0.6:
                print("The follwing image has has potential error: " + path )
            image = Image.open(path)
            image = image.crop((out_boxes[i][1], out_boxes[i][0], out_boxes[i][3], out_boxes[i][2]))
            image.save(path, "JPEG", quality=80, optimize=True, progressive=True)
            flag = True
            break
    if flag == False:
        print("The follwing image has not been cropped: " + path )
    return image
## change path with 'bird_dataset/val_images/', 'bird_dataset/train_images/' and 'bird_dataset/test_images/' and run
def save_images(model, path = 'bird_dataset/val_images/'):
    for dossier, sous_dossiers, fichiers in os.walk(path):
        if sous_dossiers == []:
            for el in fichiers:
                img = crop_images(dossier + '/' + el, model)
```
